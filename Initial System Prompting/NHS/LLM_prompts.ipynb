# Dataset Process

```python
import io

def get_df_info(df):
    output_string = ""
    # Get a concise summary of the DataFrame
    buf = io.StringIO()
    df.info(buf=buf)
    df_info = buf.getvalue()
    output_string += f"### DataFrame Info:\n{df_info}\n\n"

    # Get data types of each column
    data_types = df.dtypes
    output_string += f"### Data Types:\n{data_types}\n\n"

    # Get statistical summaries of numeric columns
    df_description = df.describe()
    output_string += f"### Statistical Summary:\n{df_description}\n\n"

    # Identify categorical columns (of type object or category)
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns

    if len(categorical_columns) != 0:
        # Check the unique values in each column to help identify categorical columns
        output_string += "### Unique Values in Categorical Columns:\n"
        for col in categorical_columns:
            unique_vals = df[col].unique()
            output_string += f"{col}: {unique_vals}\n"
        output_string += "\n"

    # Summary for each column
    summary = df.apply(lambda x: x.describe() if x.dtype in ['int64', 'float64'] else x.value_counts())
    output_string += "### Summary for Each Column:\n"
    for col, desc in summary.items():
        output_string += f"{col}:\n{desc}\n\n"
    output_string += "\n"

    # correlation
    # Get correlation matrix for numeric columns
    # Filter numeric columns for correlation matrix
    numeric_columns = df.select_dtypes(include=['number'])  # Filters for numeric columns (int64, float64)
    correlation_matrix = numeric_columns.corr()  # Compute correlation matrix for numeric columns
    output_string += "### Correlation Matrix for numeric columns:\n"
    output_string += f"{correlation_matrix}\n\n"


    return output_string


print(get_df_info(data_iris))
```

```python
def get_data_prompt(df, n_rows=10):
    df_random_sample = df.sample(n_rows)
    df_metadata = get_df_info(df)

    return f"Here is a random sample of the data: \n### DATA:\n{df_random_sample}\n\n### METADATA:\n{df_metadata}"

print(get_data_prompt(data_iris, 5))
```

```python
# Function to replace placeholder in the .md file
def replace_placeholder_in_md(file_path, placeholder, dynamic_value):
    # Open the markdown file and read its content
    with open(file_path, 'r') as file:
        content = file.read()
    
    # Replace the placeholder with the dynamic value
    updated_content = content.replace(placeholder, dynamic_value)

    # Alternatively, you can just return the updated content
    return updated_content

# Example usage
file_path = 'prompt.md'
placeholder = '{PLACEHOLDER_DATASET_METADATA}'
dynamic_value = get_data_prompt(data_iris, 10)
updated_prompt = replace_placeholder_in_md(file_path, placeholder, dynamic_value)

print(updated_prompt)
```

```python
# write string to file
def write_string_to_file(file_path, content):
    with open(file_path, 'w') as file:
        file.write(content)
```

```python
# Iris data

def get_gemini_output(df, output_text_name):
    # Example usage
    file_path = 'prompt.md'
    placeholder = '{PLACEHOLDER_DATASET_METADATA}'
    dynamic_value = get_data_prompt(df, 10)
    updated_prompt = replace_placeholder_in_md(file_path, placeholder, dynamic_value)

    print("######## Prompt in usage: ", updated_prompt)

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=[updated_prompt]
    )

    print("########Response: \n", response.text)

    output_string = f"######## Prompt in usage: {updated_prompt}\n\n"
    output_string += f"########Response: {response.text}\n\n"
    write_string_to_file(output_text_name, output_string)

get_gemini_output(data_iris, "output/gemini2_iris.txt")
```

```python
import transformers
import torch

def get_llama_output(df, output_text_name):
    # Example usage
    file_path = 'prompt.md'
    placeholder = '{PLACEHOLDER_DATASET_METADATA}'
    dynamic_value = get_data_prompt(df, 10)
    updated_prompt = replace_placeholder_in_md(file_path, placeholder, dynamic_value)

    # split the prompt into two parts
    prompt_part1 = updated_prompt.split("# User")[0]
    prompt_part2 = updated_prompt.split("# User")[1]

    print("######## System Prompt in usage: ", prompt_part1)
    print("######## User Prompt in usage: ", prompt_part2)

    model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

    pipeline = transformers.pipeline(
        "text-generation",
        model=model_id,
        model_kwargs={"torch_dtype": torch.bfloat16},
        device_map="auto",
    )

    messages = [
        {"role": "system", "content": f"{prompt_part1}"},
        {"role": "user", "content": f"{prompt_part2}"},
    ]

    outputs = pipeline(
        messages,
        max_new_tokens=512,
    )

    output = outputs[0]["generated_text"][-1]['content']


    print("########Response: \n", output)

    output_string = f"######## System Prompt in usage: {prompt_part1}\n\n"
    output_string += f"######## User Prompt in usage: {prompt_part2}\n\n"
    output_string += f"########Response: {output}\n\n"
    write_string_to_file(output_text_name, output_string)
```

```python
get_llama_output(data_iris, "output/llama_iris.txt")
get_llama_output(data_wine, "output/llama_wine.txt")
get_llama_output(data_bc, "output/llama_bc.txt")
get_llama_output(data_d, "output/llama_d.txt")
get_llama_output(data_country, "output/llama_country.txt")
```

# Qwen

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen2.5-7B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def get_qwen_output(model, tokenizer, df, output):

    # Example usage
    file_path = 'prompt.md'
    placeholder = '{PLACEHOLDER_DATASET_METADATA}'
    dynamic_value = get_data_prompt(df, 10)
    updated_prompt = replace_placeholder_in_md(file_path, placeholder, dynamic_value)

    # split the prompt into two parts
    prompt_part1 = updated_prompt.split("# User")[0]
    prompt_part2 = updated_prompt.split("# User")[1]

    print("######## System Prompt in usage: ", prompt_part1)
    print("######## User Prompt in usage: ", prompt_part2)


    messages = [
        {"role": "system", "content": prompt_part1},
        {"role": "user", "content": prompt_part2}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=512
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

    print("########Response: \n", response)
    output_string = f"######## System Prompt in usage: {prompt_part1}\n\n"
    output_string += f"######## User Prompt in usage: {prompt_part2}\n\n"
    output_string += f"########Response: {response}\n\n"
    write_string_to_file(output, output_string)
```

```python
get_qwen_output(model, tokenizer, data_iris, "output/qwen_iris.txt")
get_qwen_output(model, tokenizer, data_wine, "output/qwen_wine.txt")
get_qwen_output(model, tokenizer, data_bc, "output/qwen_bc.txt")
get_qwen_output(model, tokenizer, data_d, "output/qwen_d.txt")
get_qwen_output(model, tokenizer, data_country, "output/qwen_country.txt")
```
